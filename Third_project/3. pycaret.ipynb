
### pycaret수행하기

```python
# 필요한 모델 import
from pycaret.classification import *

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetes_data.csv")

model = setup(data=df, 
              target='Diabetes', 
              train_size=0.7, # default value
              session_id=33) # Random seed 
```

```python
# 학습 수행
top_3_models = compare_models(sort='Accuracy', # Other options are 'AUC', 'Recall', 'Precision', 'F1', 'Kappa' and 'MCC'
                              n_select = 3) # Select top n models 
```

#### 정확도가 가장 높은 세개의 모델

1. Gradient Boosting Classifier
2. Light Gradient Boosting Machine
3. Ada Boost Classifier

```python
# 정확도가 가장 높은Gradient Boosting Classifier의 평가결과
evaluate_model(top_3_models[0])
```

```python
# 정확도가 가장높은 Gradient Boosting Classifier의 예측결과
predict_model(top_3_models[0])
```

#### 모델 블렌딩
- 정확도가 가장 높은 세개의 모델을 혼합하여 더 정확도가 높은 모델을 만들기

```python
# 블렌딩된 모델 생성
blended = blend_models(estimator_list=top_3_models, 
                       fold=10, # default
                       optimize='Accuracy',
                       method = 'hard')
```

```python
# 블렌딩된 모델로 예측
predict_model(blended) 
```
